```python
class Cab:
  #initialize cab details
  def __init__(self, cab_id, driver_name, rate_per_km):
    self.cab_id = cab_id
    self.driver_name = driver_name
    self.rate_per_km = rate_per_km
    self.is_available = True

  # Show cab info
  def show_cab_info(self):
    status = "Available" if self.is_available else "Booked"
    print(f"Cab ID: {self.cab_id}")
    print(f"Driver Name: {self.driver_name}")
    print(f"Rate Per KM: {self.rate_per_km}")
    print(f"Status: {status}")


class Booking:
  def __init__(self):
    self.cabs =[
        Cab(1, "John", 10),
        Cab(2, "Alice", 15),
        Cab(3, "Bob", 20)
    ]

#Display all cabs
  def display_cabs(self):
    for cab in self.cabs:
      cab.show_cab_info()
      print()
#Book a cab
  def book_cab(self):
    cab_id = int(input("Enter cab ID to book: "))
    distance = float(input("Enter distance in KM: "))
    for cab in self.cabs:
      if cab.cab_id == cab_id:
           if cab.is_available:
               total_fare = distance * cab.rate_per_km
               cab.is_available = False
               print(f"Cab booked! Fare:₹{total_fare}")
           else:
               print(f"Cab with ID {cab_id} is not available")
           return # Exit after finding the cab
    print("Cab not found")


#cancel a booking
  def cancel_booking(self):
    cab_id = int(input("Enter cab ID to cancel booking: "))

    for cab in self.cabs:
      if cab.cab_id == cab_id:
        if not cab.is_available:
          cab.is_available = True
          print("Booking cancelled successfully")
          return
        else:
          print(f"Cab with ID {cab_id} is not booked")
          return
    print("Cab not found")

#Run CLI menu
def main():
    booking = Booking()
    while True:
      print("\n--- Cab Booking Menu ---")

      print("1. Display Cabs")
      print("2. Book Cab")
      print("3. Cancel Booking")
      print("4. Exit")

      choice = input("Enter your choice: ")
      if choice == "1":
        booking.display_cabs() # Corrected method name
      elif choice == "2":
        booking.book_cab()
      elif choice == "3":
        booking.cancel_booking()
      elif choice == "4":
        print("Thank you for using cab booking App. ")
        break
      else:
        print("Invalid choice. Try again")

# Entry point of the app

if __name__ =="__main__":
  main()
```
